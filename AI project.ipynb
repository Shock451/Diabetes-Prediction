```python
variables = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = data[variables]
y = data.Outcome
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,  y_train , y_test = train_test_split(X, y, shuffle = True)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
model = SVC(gamma='scale')
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
print ("RSME Squared" , sqrt(mean_squared_error( y_test , prediction)))
print ("RSME" , mean_squared_error( y_test , prediction))
```

```python
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import average_precision_score
```

```python
print ("Balanced Accuracy Score Squared : " , sqrt(balanced_accuracy_score( y_test , prediction)))
print ("Accuracy Score Squared : " , sqrt(accuracy_score( y_test , prediction)))
print ("Average Precision Score Squared : " , sqrt(average_precision_score( y_test , prediction)))

print ("Balanced Accuracy Score : " , balanced_accuracy_score( y_test , prediction))
print ("Accuracy Score : " , accuracy_score( y_test , prediction))
print ("Average Precision Score : " , average_precision_score( y_test , prediction))
```

# Pipeline

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
pipeline = Pipeline([
    ('nomalizer', StandardScaler()), #normalize data
    ('Models', LogisticRegression()) #use classifier
])
pipeline.steps
```

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(pipeline, X_train, y_train)
scores
```

```python
Models = []
Models.append(GradientBoostingClassifier())
Models.append(RandomForestClassifier())
Models.append(DecisionTreeClassifier())
Models.append(SVC(gamma='scale'))
Models.append(LogisticRegression())
```

```python
for classifiers in Models:
    pipeline.set_params(Models = classifiers)
    scores = cross_validate(pipeline, X_train, y_train)
    print('***************************************************************************************')
    print(str(classifiers))
    print('***************************************************************************************')
```
